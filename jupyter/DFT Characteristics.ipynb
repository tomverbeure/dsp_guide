```python
#! /usr/bin/env python3

import matplotlib.pyplot as plt
import numpy as np
from scipy.fftpack import fft, fftfreq
```

```python
def plot_time_freq(x, y, Y, N):
    freqs = fftfreq(N)
    freq_mask = freqs >= 0.0
    plt.rcParams['figure.figsize'] = [10,10]
    fig, (plt_y, plt_Y, plt_a) = plt.subplots(3)
    
    plt_y = plt.subplot(311)
    plt_y.plot(x, y, "r+")
    plt_y.grid(True)
    
    plt_Y = plt.subplot(312)
    plt_Y.plot(freqs[freq_mask] * N, np.abs(Y[freq_mask]), "r.")
    plt_Y.grid(True)
    
    plt_a = plt.subplot(313)
    phase = np.angle(Y, deg=True)
    plt_a.plot(freqs[freq_mask] * N, phase[freq_mask], "r")
    plt_a.set_ylim(-200,200)
    plt_a.grid(True)    
```

```python
N = 32

x = np.linspace(0, 1, N)
```

The DFT of an impulse has a spectrum with all amplitudes the same value.

When the input is symmetric around point 0, then the phase is 0 for all frequencies.

```python
y = np.zeros(N)
y[0] = 1


Y = fft(y)
plot_time_freq(x, y, Y, N)
```

```python
y = np.zeros(N)
y[1] = 1

Y = fft(y)
plot_time_freq(x, y, Y, N)
```

```python
y = np.zeros(N)
offset = 3

y[(-1 + offset) % N] = -5
y[( 0 + offset) % N] = 10
y[( 1 + offset) % N] = -5

Y = fft(y)
plot_time_freq(x, y, Y, N)
```
