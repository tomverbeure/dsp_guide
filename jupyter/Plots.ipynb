Here's a good overall tutorial: https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 10)
```

When used with a single range, `plot` uses the given values as Y, and X is simply numbers from 0 to the number of values.

In this case, we're going from 0 to 2 * Pi:

```python
plt.plot(x)
# show() isn't strictly necessary for the graph to show up in a Jupyter notebook, but since
# most pyplot command return an object or data of some sort, that output will end up in the result.
# show() doesn't have an output and that makes things clean.
plt.show()
```

```python
plt.plot(x)
plt.ylabel("Number from the X array")
plt.show()
```

When there are 2 arguments, the first argument is an array with X values, and the second one are the corresponding Y values:

```python
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

The third parameter says how to mark things:
    

```python
plt.plot(x,y, "r+")
plt.show()
```

You can give a sequence of 3-parameter sets:

```python
plt.plot(x,y, "r+", x, 2*y, "g*")
plt.show()
```

You can even mix 2 and 4 parameter sets, but this can be confusing:
* the first 3 parameters have an X axis range from 0 to 2*PI
* the second 2 parameters have an X axis range from 0 to 9, beause there ae 10 Y values.

```python
plt.plot(x,y, "r+", 2*y, "g*")
plt.show()
```

Multiple planes:

```python
# Overall plot size is 9 wide and 3 high
plt.figure(figsize=(9,6))

# There are 2 rows and 3 columns -> 6 graphs
# Use the last number to address the 
plt.subplot(231)
plt.plot([1], "r*")
plt.subplot(232)
plt.plot([1,2], "r*")
plt.subplot(233)
plt.plot([1,2,3], "r*")
plt.subplot(234)
plt.plot([1,2,3,4], "r*")
plt.subplot(235)
plt.plot([1,2,3,4,5], "r*")
plt.subplot(236)
plt.plot([1,2,3,4,5,6], "r*")
plt.show()
```

Multiple figures:

```python
plt.figure(1, figsize=(6,6))
plt.subplot(121)
plt.plot([1,2,3,4])
plt.subplot(122)
plt.plot([4,3,2,1])

plt.figure(2, figsize=(4,4))
plt.plot([1,2,3,4])
plt.show()
```

Subplots, with shared X and Y

```python
fig, axes = plt.subplots(3,2, sharex=True, sharey=True)
plt.rcParams['figure.figsize'] = [10,10]
axes[0,0].plot([1],[1], "r+")
axes[2,1].plot([6],[6], "r+")
```
